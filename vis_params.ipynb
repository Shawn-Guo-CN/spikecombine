```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import multivariate_normal
from mpl_toolkits.mplot3d import Axes3D
# %matplotlib inline  
```

```python
from spikecombine import DataGenerator
from spikecombine import DataLoader
from spikecombine import SpikeSortersCombiner
```

```python
ssc = SpikeSortersCombiner()
ssc.load('e:\\Edin\\RA\\projects\\spikecombine\\data\\test3.params')
```

```python
sorter_params = ssc.get_params_for_sorter('tridesclous')

mu1 = sorter_params['positive_mean']
covar1 = sorter_params['positive_covar']
rv1 = multivariate_normal(mu1, covar1)

mu2 = sorter_params['negative_mean']
covar2 = sorter_params['negative_covar']
rv2 =  multivariate_normal(mu2, covar2)
```

```python
#Create grid and multivariate normal
x = np.linspace(-1200,1200,50)
y = np.linspace(-12000,12000,50)

X, Y = np.meshgrid(x,y)
pos = np.empty(X.shape + (2,))
pos[:, :, 0] = X; pos[:, :, 1] = Y
```

```python
#Make a 3D plot
fig = plt.figure()
ax = fig.gca(projection='3d')
ax.plot_surface(X, Y, rv1.pdf(pos),cmap='viridis',linewidth=0)
ax.set_xlabel('X axis')
ax.set_ylabel('Y axis')
ax.set_zlabel('Z axis')
plt.show()
```

```python
fig = plt.figure()
ax = fig.gca(projection='3d')
ax.plot_surface(X, Y, rv2.pdf(pos),cmap='viridis',linewidth=0)
ax.set_xlabel('X axis')
ax.set_ylabel('Y axis')
ax.set_zlabel('Z axis')
plt.show()
```
